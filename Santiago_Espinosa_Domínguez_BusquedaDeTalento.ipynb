## 2. Generación de CVs sintéticos

Para esta actividad se generaron 20 currículums sintéticos utilizando inteligencia artificial. Debido a que la actividad busca simular un sistema de búsqueda de talento sin comprometer información personal real, los CVs no corresponden a personas reales.

Cada CV incluye información profesional relevante para el análisis de talento, como resumen profesional, formación académica, experiencia laboral, habilidades técnicas, habilidades blandas, certificaciones, idiomas y proyectos relevantes.

Los perfiles fueron generados con tres niveles de experiencia:

- **Junior:** perfiles con poca experiencia profesional, normalmente de 0 a 2 años. Suelen contar con conocimientos básicos, proyectos académicos o prácticas profesionales.
- **Intermedio o Semi-Senior:** perfiles con experiencia media, normalmente de 3 a 5 años. Cuentan con experiencia laboral comprobable, participación en proyectos reales y dominio de herramientas técnicas.
- **Senior:** perfiles con experiencia avanzada, normalmente de más de 6 años. Suelen liderar proyectos, diseñar arquitecturas, tomar decisiones técnicas y coordinar equipos.

## 3. Importancia de no incluir datos personales sensibles

En la generación de los CVs sintéticos se evitó incluir información como nombre completo, fotografía, género, edad, nacionalidad, estado civil, domicilio o cualquier otro dato personal sensible.

Esto es importante porque un sistema de búsqueda de talento debe enfocarse en las competencias, experiencia, formación, habilidades y proyectos de las personas candidatas, no en características personales que puedan generar sesgos o discriminación.

Excluir este tipo de información ayuda a que el sistema sea más justo, ético y objetivo. Además, permite proteger la privacidad de las personas y reducir el riesgo de que el modelo favorezca o descarte candidatos por razones no relacionadas con su capacidad profesional.

## 6. Generación de archivos PDF y PNG de los CVs sintéticos

Para simular una base de conocimiento multimodal, los CVs sintéticos se guardaron en dos formatos diferentes: PDF y PNG. Esto permite que el sistema inteligente trabaje con documentos de texto y también con imágenes, como ocurriría en un proceso real de reclutamiento donde los CVs pueden recibirse en distintos formatos.

La distribución utilizada fue la siguiente:

- 1 CV Junior en formato PNG.
- 2 CVs Intermedio/Semi-Senior en formato PNG.
- 1 CV Senior en formato PNG.
- Los 16 CVs restantes en formato PDF.

Esta variedad permite probar la capacidad del sistema para procesar información proveniente de diferentes tipos de archivo.

```python
# Importación de librerías para generación de archivos

import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from PIL import Image, ImageDraw, ImageFont
import textwrap
```

```python
# Creación de carpeta para almacenar los CVs sintéticos

carpeta_cvs = "cvs_sinteticos"

os.makedirs(carpeta_cvs, exist_ok=True)

print("Carpeta creada:", carpeta_cvs)
```

```python
# Función para guardar un CV en formato PDF

def guardar_cv_pdf(cv, carpeta):
    """
    Guarda el contenido de un CV sintético en formato PDF.
    """

    ruta_archivo = os.path.join(carpeta, f"{cv['id']}.pdf")

    c = canvas.Canvas(ruta_archivo, pagesize=letter)
    width, height = letter

    x = 50
    y = height - 50

    # Título del CV
    c.setFont("Helvetica-Bold", 14)
    c.drawString(x, y, f"{cv['id']} - {cv['nivel']} - {cv['perfil']}")

    y -= 30

    # Contenido del CV
    c.setFont("Helvetica", 10)

    lineas = cv["contenido"].split("\n")

    for linea in lineas:
        linea = linea.strip()

        if linea == "":
            y -= 10
            continue

        lineas_ajustadas = textwrap.wrap(linea, width=90)

        for linea_ajustada in lineas_ajustadas:
            if y < 50:
                c.showPage()
                c.setFont("Helvetica", 10)
                y = height - 50

            c.drawString(x, y, linea_ajustada)
            y -= 14

    c.save()

    return ruta_archivo
```

```python
# Función para guardar un CV en formato PNG

def guardar_cv_png(cv, carpeta):
    """
    Guarda el contenido de un CV sintético en formato PNG.
    """

    ruta_archivo = os.path.join(carpeta, f"{cv['id']}.png")

    ancho = 1200
    alto = 1600

    imagen = Image.new("RGB", (ancho, alto), "white")
    draw = ImageDraw.Draw(imagen)

    try:
        fuente_titulo = ImageFont.truetype("DejaVuSans-Bold.ttf", 28)
        fuente_texto = ImageFont.truetype("DejaVuSans.ttf", 22)
    except:
        fuente_titulo = ImageFont.load_default()
        fuente_texto = ImageFont.load_default()

    x = 50
    y = 50

    # Título del CV
    draw.text(
        (x, y),
        f"{cv['id']} - {cv['nivel']} - {cv['perfil']}",
        fill="black",
        font=fuente_titulo
    )

    y += 50

    # Contenido del CV
    lineas = cv["contenido"].split("\n")

    for linea in lineas:
        linea = linea.strip()

        if linea == "":
            y += 15
            continue

        lineas_ajustadas = textwrap.wrap(linea, width=85)

        for linea_ajustada in lineas_ajustadas:
            draw.text(
                (x, y),
                linea_ajustada,
                fill="black",
                font=fuente_texto
            )
            y += 32

            if y > alto - 80:
                break

    imagen.save(ruta_archivo)

    return ruta_archivo
```

```python
# Selección de CVs que se guardarán como PNG

cvs_png = ["CV_01", "CV_06", "CV_11", "CV_16"]
print("CVs que se guardarán como PNG:", cvs_png)
```

```python
# Generación de archivos PDF y PNG

archivos_generados = []
for cv in cvs_sinteticos:
    if cv["id"] in cvs_png:
        ruta = guardar_cv_png(cv, carpeta_cvs)
        formato = "PNG"
    else:
        ruta = guardar_cv_pdf(cv, carpeta_cvs)
        formato = "PDF"

    archivos_generados.append({
        "id": cv["id"],
        "nivel": cv["nivel"],
        "perfil": cv["perfil"],
        "formato": formato,
        "ruta": ruta
    })

df_archivos = pd.DataFrame(archivos_generados)

df_archivos
```

## 7. Sistema inteligente multimodal para búsqueda de talento

En esta sección se construye un sistema inteligente para recuperar y recomendar candidatos a partir de una base de conocimiento formada por CVs sintéticos en formato PDF y PNG.

El sistema realiza las siguientes etapas:

1. Extracción de texto desde archivos PDF y PNG.
2. Construcción de una representación textual de la vacante.
3. Generación de embeddings para la vacante y para cada CV.
4. Construcción de una base vectorial.
5. Búsqueda semántica de los candidatos más similares a la vacante.
6. Selección y justificación de los 5 mejores candidatos.

Para los archivos PDF se extrae texto directamente del documento. Para los archivos PNG se utiliza OCR, ya que el texto se encuentra dentro de una imagen.

```python
# Instalación de librerías necesarias

!apt-get install -y tesseract-ocr
!pip install pytesseract pypdf sentence-transformers faiss-cpu
```

```python
# Importación de librerías

import os
import numpy as np
import pandas as pd
import pytesseract
import faiss

from PIL import Image
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
```

```python
# Función para extraer texto de archivos PDF

def extraer_texto_pdf(ruta_archivo):
    texto = ""
    reader = PdfReader(ruta_archivo)
    for page in reader.pages:
        texto_pagina = page.extract_text()
        if texto_pagina:
            texto += texto_pagina + "\n"
    return texto
```

```python
# Función para extraer texto de archivos PNG mediante OCR

def extraer_texto_png(ruta_archivo):
    imagen = Image.open(ruta_archivo)
    texto = pytesseract.image_to_string(imagen, lang="eng")
    return texto
```

```python
# Extracción de texto de todos los CVs generados

textos_cvs = []

for _, row in df_archivos.iterrows():
    ruta = row["ruta"]
    formato = row["formato"]
    if formato == "PDF":
        texto = extraer_texto_pdf(ruta)
    elif formato == "PNG":
        texto = extraer_texto_png(ruta)
    else:
        texto = ""

    textos_cvs.append({
        "id": row["id"],
        "nivel": row["nivel"],
        "perfil": row["perfil"],
        "formato": formato,
        "ruta": ruta,
        "texto_extraido": texto
    })

df_textos_cvs = pd.DataFrame(textos_cvs)
df_textos_cvs.head()
```

```python
# Revisión rápida del texto extraído de un CV

print(df_textos_cvs.loc[0, "texto_extraido"][:1000])
```

La extracción de texto permite convertir los archivos PDF y PNG en información procesable por el sistema. Esto es necesario porque los modelos de búsqueda semántica no trabajan directamente con documentos visuales, sino con representaciones textuales.

En el caso de los archivos PDF, el texto se extrae directamente. En el caso de los archivos PNG, se utiliza OCR para reconocer el contenido escrito dentro de la imagen.

```python
# Construcción del texto de la vacante

texto_vacante = f"""
Empresa: {vacante['empresa']}

Nombre de la vacante: {vacante['nombre_vacante']}

Descripción:
{vacante['descripcion']}

Competencias requeridas:
{', '.join(vacante['competencias'])}

Responsabilidades:
{', '.join(vacante['responsabilidades'])}

Requisitos técnicos:
{', '.join(vacante['requisitos_tecnicos'])}
"""

print(texto_vacante)
```

```python
# Carga del modelo de embeddings

modelo_embeddings = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
# Generación de embeddings para los CVs

lista_textos_cvs = df_textos_cvs["texto_extraido"].tolist()

embeddings_cvs = modelo_embeddings.encode(
    lista_textos_cvs,
    convert_to_numpy=True,
    normalize_embeddings=True
)
embeddings_cvs.shape
```

```python
# Generación del embedding de la vacante

embedding_vacante = modelo_embeddings.encode(
    [texto_vacante],
    convert_to_numpy=True,
    normalize_embeddings=True
)

embedding_vacante.shape
```

```python
# Construcción de la base de datos vectorial con FAISS

dimension = embeddings_cvs.shape[1]
index = faiss.IndexFlatIP(dimension)
index.add(embeddings_cvs)
print("Cantidad de CVs en la base vectorial:", index.ntotal)
```

Para construir la base vectorial se utilizó FAISS, una biblioteca especializada en búsqueda eficiente de vectores. Cada CV fue convertido en un embedding, es decir, una representación numérica de su contenido semántico.

También se generó un embedding para la vacante seleccionada. Posteriormente, la búsqueda se realiza comparando la similitud entre el vector de la vacante y los vectores de los CVs.

En este caso se utiliza similitud por producto interno sobre embeddings normalizados, lo cual equivale a una comparación basada en similitud coseno.

```python
# Búsqueda semántica de los 5 mejores candidatos

top_k = 5
scores, indices = index.search(embedding_vacante, top_k)
resultados_recomendacion = []
for rank, idx in enumerate(indices[0]):
    candidato = df_textos_cvs.iloc[idx]
    resultados_recomendacion.append({
        "ranking": rank + 1,
        "id": candidato["id"],
        "nivel": candidato["nivel"],
        "perfil": candidato["perfil"],
        "formato": candidato["formato"],
        "score_similitud": float(scores[0][rank]),
        "ruta": candidato["ruta"],
        "texto_extraido": candidato["texto_extraido"]
    })

df_top5 = pd.DataFrame(resultados_recomendacion)
df_top5[["ranking", "id", "nivel", "perfil", "formato", "score_similitud"]]
```

```python
# Función para generar justificación básica de la recomendación

def generar_justificacion(texto_cv, perfil, nivel):
    texto_lower = texto_cv.lower()

    criterios = {
        "Python": "python" in texto_lower,
        "Machine Learning": "machine learning" in texto_lower or "aprendizaje automático" in texto_lower,
        "SQL": "sql" in texto_lower,
        "NLP": "nlp" in texto_lower or "procesamiento de lenguaje natural" in texto_lower,
        "PySpark/Spark": "pyspark" in texto_lower or "spark" in texto_lower,
        "Cloud": "aws" in texto_lower or "azure" in texto_lower or "cloud" in texto_lower,
        "Embeddings": "embedding" in texto_lower or "embeddings" in texto_lower,
        "MLOps": "mlops" in texto_lower,
        "Modelos predictivos": "modelo" in texto_lower or "modelos predictivos" in texto_lower
    }

    coincidencias = [k for k, v in criterios.items() if v]

    justificacion = f"""
    El candidato presenta un perfil de tipo {perfil} con nivel {nivel}.
    Su CV muestra coincidencias relevantes con la vacante en áreas como:
    {', '.join(coincidencias)}.
    Por esta razón, el sistema lo considera un perfil cercano a los requisitos
    técnicos y responsabilidades del puesto.
    """

    return justificacion.strip()
```

```python
# Agregar justificación a los 5 mejores candidatos

df_top5["justificacion"] = df_top5.apply(
    lambda row: generar_justificacion(
        row["texto_extraido"],
        row["perfil"],
        row["nivel"]
    ),
    axis=1
)

df_top5[["ranking", "id", "nivel", "perfil", "score_similitud", "justificacion"]]
```

## 8. Reporte técnico de la arquitectura implementada

En esta actividad se diseñó un sistema inteligente multimodal para búsqueda de talento. La arquitectura implementada inicia con la generación de una base de conocimiento formada por 20 CVs sintéticos en formato PDF y PNG. Posteriormente, se extrae el texto de cada documento: para los archivos PDF se utiliza extracción directa de texto, mientras que para los archivos PNG se utiliza OCR.

Después de obtener el texto de cada CV, se construye una representación textual de la vacante seleccionada. Tanto los CVs como la vacante son convertidos en embeddings utilizando un modelo multilingüe de Sentence Transformers. Estos embeddings permiten representar el significado semántico de los documentos en forma vectorial.

Con los embeddings de los CVs se construye una base vectorial utilizando FAISS. Luego, el embedding de la vacante se compara contra la base vectorial para recuperar los candidatos con mayor similitud semántica. Finalmente, el sistema selecciona los 5 mejores candidatos y genera una justificación basada en la coincidencia entre sus habilidades, experiencia y los requisitos de la vacante.

La arquitectura general del sistema puede resumirse en las siguientes etapas:

1. Selección de vacante real.
